```python
def spearmanr_pval(x,y):
    return spearmanr(x,y)[1]

def calculate_pvalues(df9):
    df9 = df9.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df9.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df9.columns:
        for c in df9.columns:
            pvalues[r][c] = round(spearmanr(df9[r], df9[c])[1], 4)
    return pvalues

prob0=calculate_pvalues(data_raw[["White","Black","Asian","Hispanic","Other","NH"]])
print(prob0)
prob1=pd.DataFrame(prob0).fillna(0)
prob=data_raw[["White","Black","Asian","Hispanic","Other","NH"]].corr(method=spearmanr_pval)#prob1.values                                                                                                       
plt.figure(figsize=(10,10))
sns.heatmap(prob, vmax=1, square=True,annot=True,cmap='Greys')
plt.title('p-values')
plt.savefig("final_images/prob_all_sp_NB.JPG", dpi=300,format="JPG")
plt.show()
```

```python
def getExtremePoints(data, typeOfExtreme = None, maxPoints = None):
    """
    This method returns the indeces where there is a change in the trend of the input series.
    typeOfExtreme = None returns all extreme points, max only maximum values and min
    only min,
    """
    a = np.diff(data)
    asign = np.sign(a)
    signchange = ((np.roll(asign, 1) - asign) != 0).astype(int)
    idx = np.where(signchange ==1)[0]
    if typeOfInflexion == 'max' and data[idx[0]] < data[idx[1]]:
        idx = idx[1:][::2]
    elif typeOfInflexion == 'min' and data[idx[0]] > data[idx[1]]:
        idx = idx[1:][::2]
    elif typeOfInflexion is not None:
        idx = idx[::2]
    
    # sort ids by min value
    if 0 in idx:
        idx = np.delete(idx, 0)
    if (len(data)-1) in idx:
        idx = np.delete(idx, len(data)-1)
    idx = idx[np.argsort(data[idx])]
    # If we have maxpoints we want to make sure the timeseries has a cutpoint
    # in each segment, not all on a small interval
    if maxPoints is not None:
        idx= idx[:maxPoints]
        if len(idx) < maxPoints:
            return (np.arange(maxPoints) + 1) * (len(data)//(maxPoints + 1))
    
    return idx
print(getExtremePoints(data_entropy[data_entropy["labels"]==2]["entropy"]))
```
